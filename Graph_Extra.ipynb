```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
cointag = pd.read_csv("coin_tags.csv")
coin = pd.read_csv("coins200.csv")
information = pd.read_csv("coins_information.csv")
tags = pd.read_csv("tags.csv")
```

```python
information['pricedif'] = information.apply( lambda row : row.close - row.open , axis=1 )
# information['mean'] = information.apply( lambda row : row.close + row.open / 2 , axis=1 )
# result_df = pd.DataFrame(columns=['Symbol 1', 'Symbol 2', 'Count'])
```

```python
df = information[['Symbol', 'timestamp', 'pricedif', "open"]].copy()
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.rename(columns = {'timestamp':'time'}, inplace = True)
df
```

```python
# 0 <= i < 200   =>   symbol_list[i]+
# 200 <= i < 400   =>   symbol_list[i]-
max_n = 20
symbol_list = df['Symbol'].copy().unique()
symbol_list = symbol_list[:max_n]
# symbol_list.sort()
symbol_list
```

```python
date_list = df['time'].copy().unique().tolist()
date_list.sort()
date_list
```

```python
dic = {}
for i in symbol_list:
    dic[i] = df[df["Symbol"] == i]
dic[symbol_list[0]]
```

```python
n = len(symbol_list)
mat = [None] * n * 2
for i in range(n * 2):
    mat[i] = [0] * n * 2
```

```python
for i in range(max_n):
    for j in range(i+1, max_n):
        a = dic[symbol_list[i]]
        b = dic[symbol_list[j]]
        for day in date_list:
            try:
                a_opn = a[a["time"]==day]["open"].values[0]
                b_opn = b[b["time"]==day]["open"].values[0]
                a_dif = a[a["time"]==day]["pricedif"].values[0]
                b_dif = b[b["time"]==day]["pricedif"].values[0]
                if(a_opn == 0):
                    a_dif = 0.000001
                if(b_opn == 0):
                    b_per = 0.000001
                a_per = a_dif / a_opn
                b_per = b_dif / b_opn
                
                percent = min(a_per,b_per) / max(a_per,b_per)

                if a_per > 0 and b_per > 0:
                    mat[i][j] += percent
                    mat[i][j+n] -= percent
                if a_per > 0 and b_per < 0:
                    mat[i][j+n] += percent
                    mat[i][j] -= percent
                if a_per < 0 and b_per > 0:
                    mat[i+n][j] += percent
                    mat[i+n][j+n] -= percent
                if a_per < 0 and b_per < 0:
                    mat[i+n][j+n] += percent
                    mat[i+n][j] -= percent
            except:
                continue
```

```python
max_mat = max(max(mat))
min_mat = min(min(mat))
max_n_go = max_n * 2
for i in range(max_n_go):
    for j in range(max_n_go):
        mat[i][j] /= (max_mat - min_mat)
```

```python
for i in range(max_n_go):
    for j in range(max_n_go):
        mat[j][i] = mat[i][j]
```

```python
for i in range(max_n_go):
    for j in range(max_n_go):
        print('%.5f'%(mat[i][j]), end=" ")
    print("")
```
