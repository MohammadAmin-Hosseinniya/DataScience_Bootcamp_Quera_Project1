Making table Coins_information

```python
coins_information = pd.DataFrame()

for i in range(coins.shape[0]):
    coin_name = coins["Name"][i]
    tail_of_csv_file_name = "_9_3_2022-9_3_2023_historical_data_coinmarketcap.csv"
    csv_name = "csvs\\" + coin_name + tail_of_csv_file_name 
    coin_detail = pd.read_csv(csv_name , delimiter=";")
    #adding symbol 
    symbol = coins["Symbol"][i]
    symbol_col =[symbol for j in range(coin_detail.shape[0])] 
    coin_detail.insert(loc = 0,
                       column = "Symbol",
                       value = symbol_col)
    coin_detail["circulatingSupply"] =coin_detail["marketCap"]/coin_detail["close"]
    coins_information = pd.concat([coins_information ,coin_detail])


coins_information["circulatingSupply"] =coins_information["marketCap"]/coins_information["close"]
coins_information
```

Making table Coins200

```python
Coins200 = coins[["Rank","Name","Symbol","MainLink","HistoricalLink","Git_hub"]]
Coins200
```

Making table Tags

```python
Tags = pd.read_csv('Tags.csv')
Tags
```

Making table coin_tags

```python
coin_tags = pd.DataFrame()
for i in range(coins.shape[0]):
    symbol = coins["Symbol"][i]
    #print(type(coins["Tags"][i]))
    try:
        tags = (coins["Tags"][i]).split(",")
        num_tags = len(tags)
        symbol_col = [symbol for j in range(num_tags)]
        tags_col = list()
        for tag in tags :
            temp = Tags[Tags["Tags"]==tag]
            temp = temp["Tags_Id"]
            tag_id = temp.iloc[0]
            tags_col.append(tag_id)
    except :
        tags_col = ["nan"]
        symbol_col = [symbol]
        

    
    data = {"Symbol":symbol_col,"Tag_Id":tags_col}
    coin_tag = pd.DataFrame(data= data)
    coin_tags = pd.concat([coin_tags,coin_tag])
        
```
